
<span style="color:Green;">Presenting the correlation between all the columns i used via a heatmap</span>

```python
selected_columns = ['rate', 'votes', 'online_order', 'book_table', 'delivery']
correlation_data = india[selected_columns].copy()

# Convert 'online_order' and 'book_table' to numeric values
correlation_data['online_order'] = correlation_data['online_order'].map({'Yes': 1, 'No': 0})
correlation_data['book_table'] = correlation_data['book_table'].map({'Yes': 1, 'No': 0})

correlation_matrix = correlation_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='GnBu', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
